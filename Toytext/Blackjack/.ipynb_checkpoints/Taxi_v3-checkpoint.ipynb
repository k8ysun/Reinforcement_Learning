```python
import gym 
import numpy as np 
from tqdm import tqdm 
```

```python
class AgentBlackJack():
    def __init__(self, epsilon):
        self.state_value = {}
        self.state_cache = []
        self.epsilon = epsilon
        

    
    def state_initialize(self, cnt_win = 10, cnt_tot = 10):
        return [[cnt_win, cnt_tot], [cnt_win,cnt_tot]]
    
    def update_state(self, reward):
        if reward == 1:
            win_delta = 1 
        else:
            win_delta = 0 
        
        for state, action in self.state_cache:
            self.state_value[state][action][0] += win_delta
            self.state_value[state][action][1] += 1
        
        self.state_cache = []
            
    
    def select_action(self, state):
        # initialize the value of the state if it has not been visited before
        if state not in self.state_value:
            self.state_value[state] = self.state_initialize()
        
        value = self.state_value[state]
        
        # with probability of epsilon, we will explore the action of the state randomly
        r = np.random.uniform(0, 1)
        if r < self.epsilon:
            r_action = np.random.uniform(0, 1)
            action = int(r_action <= 0.5)

        # pick the state with the larger value
        elif value[0][0] / value[0][1] <= value[1][0] / value[1][1]:
            action = 1
        else:
            action = 0 
        
        # save the state action pair in cache
        self.state_cache.append((state, action))
        
        return action
        
        
```
